```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import json

# Configure Chrome options
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--headless=new")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

collections = ['RADIOMIR']
base_url = "https://www.panerai.com/fr/fr/collections/watch-collection/{}.html"

def extract_price_data(price_str):
    """Handle price formatting with non-breaking spaces"""
    if price_str == "N/A":
        return "N/A", "N/A"
    parts = price_str.replace('\xa0', ' ').split()
    return ' '.join(parts[:-1]), parts[-1]

for collection in collections:
    try:
        url = base_url.format(collection.lower())
        driver.get(url)
        print(f"Scraping: {url}")
        
        # Wait for main container instead of individual elements
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, ".rcms_eelistview"))
        )
        
        # More specific selector to avoid duplicates
        product_cards = driver.find_elements(
            By.CSS_SELECTOR, "div.pan-prod-ref-wrapper-v2.rcms_eelistview > div.pan-prod-ref-card-v2")
        print(f"Found {len(product_cards)} products for collection: {collection}")

        for card in product_cards:
            try:
                # Get the main product link container
                product_link_element = WebDriverWait(card, 5).until(
                    EC.presence_of_element_located((By.CLASS_NAME, "pan-prod-ref-link-v2"))
                )
                
                # Extract JSON data
                data_tracking = json.loads(
                    product_link_element.get_attribute("data-tracking-product")
                    .replace("&quot;", '"')
                )

                # Extract basic info from JSON
                product_info = {
                    'name': data_tracking.get('name', 'N/A'),
                    'reference': data_tracking.get('reference', 'N/A'),
                    'collection': data_tracking.get('collection', 'N/A'),
                    'brand': data_tracking.get('brand', 'N/A'),
                    'price': data_tracking.get('price', 'N/A'),
                    'availability': "Available" if data_tracking.get('isAvailable', 'false') == 'true' else "Out of Stock",
                    'url': product_link_element.get_attribute('href')
                }

                # Handle price and currency
                product_info['price'], product_info['currency'] = extract_price_data(product_info['price'])

                # Image extraction
                try:
                    main_img = card.find_element(
                        By.CSS_SELECTOR, "picture.pan-picture-tag source[media='(min-width: 1000px)']"
                    ).get_attribute("srcset").split()[0]
                except:
                    main_img = 'N/A'

                try:
                    night_img = card.find_element(
                        By.CSS_SELECTOR, ".pan-prod-ref-night-image-v2 source[media='(min-width: 1000px)']"
                    ).get_attribute("srcset").split()[0]
                except:
                    night_img = 'N/A'

                product_info.update({
                    'main_image': main_img,
                    'night_image': night_img
                })

                # Print results
                for key, value in product_info.items():
                    print(f"{key.replace('_', ' ').title()}: {value}")
                print("-" * 50)

            except Exception as e:
                print(f"Error processing product: {str(e)}")
                continue

    except Exception as e:
        print(f"Error processing {collection}: {str(e)}")
        continue

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import json
import time

# Configure Chrome options
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--headless=new")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

collections = ['RADIOMIR']
base_url = "https://www.panerai.com/fr/fr/collections/watch-collection/{}.html"

# Function to extract product data from a product card
def extract_product_data(product_card):
    try:
        # Extract product details from the data-tracking-product attribute
        tracking_data = json.loads(
            product_card.find_element(By.CSS_SELECTOR, "a.pan-prod-ref-link-v2").get_attribute("data-tracking-product")
        )
        
        # Extract image URLs
        image_url = product_card.find_element(
            By.CSS_SELECTOR, "img.pan-adaptive-image-util"
        ).get_attribute("src")
        
        # Extract price and tax information
        price_element = WebDriverWait(product_card, 5).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "p.product-price"))
        )
        price = price_element.text.replace(" ", " ")  # Replace non-breaking spaces
        
        tax_info = product_card.find_element(By.CSS_SELECTOR, "p.product-tax").text
        
        # Extract product name and reference
        product_name = product_card.find_element(By.CSS_SELECTOR, "p.product-name").text
        product_reference = product_card.find_element(By.CSS_SELECTOR, "p.product-reference").text
        
        # Build the product JSON
        product_json = {
            "id": tracking_data.get("id"),
            "name": product_name,
            "brand": tracking_data.get("brand"),
            "category": tracking_data.get("category"),
            "collection": tracking_data.get("collection"),
            "reference": product_reference,
            "price": {
                "amount": float(tracking_data.get("price").replace(" ", "").replace("€", "").strip()),
                "currency": tracking_data.get("currency"),
                "display": price,
                "tax_status": tax_info
            },
            "details": {
                "size": tracking_data.get("size"),
                "variant": tracking_data.get("variant"),
                "material": {
                    "case": tracking_data.get("productMaterialCase"),
                    "strap": tracking_data.get("productMaterialStrap")
                },
                "special_edition": tracking_data.get("productSpecialEdition") == "true"
            },
            "availability": {
                "is_available": tracking_data.get("isAvailable") == "true",
                "is_sellable": tracking_data.get("isSellable") == "true"
            },
            "url": product_card.find_element(By.CSS_SELECTOR, "a.pan-prod-ref-link-v2").get_attribute("href"),
            "images": {
                "main": image_url
            }
        }
        return product_json
    except Exception as e:
        print(f"Error extracting product data: {e}")
        return None

# Function to scrape a collection
def scrape_collection(collection_name):
    try:
        # Navigate to the collection page
        url = base_url.format(collection_name.lower())
        driver.get(url)
        
        # Wait for the product cards to load
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "div.pan-prod-ref-card-v2"))
        )
        
        # Scroll to load all products (if lazy loading is used)
        last_height = driver.execute_script("return document.body.scrollHeight")
        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)  # Wait for new content to load
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height
        
        # Extract all product cards
        product_cards = driver.find_elements(By.CSS_SELECTOR, "div.pan-prod-ref-card-v2")
        
        # Extract data for each product
        products = []
        for card in product_cards:
            product_data = extract_product_data(card)
            if product_data:
                products.append(product_data)
        
        # Save the data to a JSON file
        output_file = f"{collection_name.lower()}_products.json"
        with open(output_file, "w", encoding="utf-8") as f:
            json.dump(products, f, indent=4, ensure_ascii=False)
        
        print(f"Scraped {len(products)} products from {collection_name} collection. Saved to {output_file}.")
    
    except Exception as e:
        print(f"Error scraping {collection_name} collection: {e}")

# Scrape each collection
for collection in collections:
    scrape_collection(collection)

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import json
import pandas as pd

# Configure Chrome options
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--headless=new")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")

# Automatic driver management
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

# collections = ['RADIOMIR', 'LUMINOR', 'SUBMERSIBLE', 'LUMINOR-DUE']
countries = {"USA": "us/en", "France": "fr/fr" , "UK": "gb/en", "Japan":"jp/ja"}
collections = ['RADIOMIR']
# countries = {"USA": "us/en"}
base_url = "https://www.panerai.com/{}/collections/watch-collection/{}.html"

for country, country_url in countries.items():
    country_products = []
    for collection in collections:        
        try:
            # Convert to lowercase and format URL
            collection_lower = collection.lower()
            url = base_url.format(country_url, collection_lower)
            driver.get(url)
            print(f"Scraping: {url}")

            # Wait for product cards to load
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CLASS_NAME, "pan-prod-ref-wrapper-v2"))
            )

            # Find all product cards
            product_cards = driver.find_elements(By.CLASS_NAME, "pan-prod-ref-wrapper-v2")
            print(f"Found {len(product_cards)} products for collection: {collection}")
            i=0
            for card in product_cards:
                if i >= len(product_cards)/2:
                    break
                try:
                    # Extract data from the data-tracking-product attribute
                    product_link_element = card.find_element(By.CLASS_NAME, "pan-prod-ref-link-v2")
                    data_tracking = product_link_element.get_attribute("data-tracking-product")
                    data_tracking = json.loads(data_tracking.replace("&quot;", '"'))

                    # Extract product details
                    product_info = {
                        'name': data_tracking.get('name', 'N/A'),
                        'reference': data_tracking.get('reference', 'N/A'),
                        'collection': data_tracking.get('collection', 'N/A'),
                        'brand': data_tracking.get('brand', 'N/A'),
                        'price': data_tracking.get('price', 'N/A'),
                        'currency': data_tracking.get('currency', 'N/A'),
                        'availability': "Available" if data_tracking.get('isAvailable', 'false') == 'true' else "Out of Stock",
                        'product_url': product_link_element.get_attribute('href')
                    }

                    # Extract main image URL with better handling
                    try:
                        img_element = WebDriverWait(card, 5).until(
                            EC.presence_of_element_located((By.CSS_SELECTOR, ".pan-prod-ref-front-image-v2 img"))
                        )
                        driver.execute_script("arguments[0].scrollIntoView();", img_element)
                        main_image = img_element.get_attribute("data-src") or img_element.get_attribute("src")
                        if "transform" in main_image:
                            main_image = main_image.split(".transform")[0]  # Remove transform part
                        product_info['image_url'] = "https://www.panerai.com" + main_image
                    except Exception as e:
                        print(f"Error extracting main image: {e}")
                        product_info['image_url'] = "N/A"

                    # Print formatted output
                    for key, value in product_info.items():
                        print(f"{key.replace('_', ' ').title()}: {value}")
                    print("-" * 50)

                except Exception as e:
                    print(f"Error processing product: {str(e)}")
                    continue
                i+=1
        
            country_products.append(product_info)
        
        except Exception as e:
            print(f"Error processing {collection}: {str(e)}")
            continue
    
    df = pd.DataFrame(country_products)
    df.to_csv(f"{country}_products_2025.csv", index=False)

driver.quit()
```
